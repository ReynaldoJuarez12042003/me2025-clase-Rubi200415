```python
def f(x):
  return np.exp(-x)-x

def g1(x):
  return np.exp(-x)

def fp(x):
  return -np.exp(-x)-1

def fpp(x):
  return np.exp(-x)

def puntofijo(gx,xi,nit,eps_max):
  for i in range(nit):
    xi1=gx(xi)
    Ea=np.abs(xi1-xi)
    print(i,xi1,Ea)
    if Ea<eps_max: break
    xi=xi1
  return


def NewtonRaphson(fx,fpx,xi,nit,eps_max):
  for i in range(nit):
    xi1=xi-fx(xi)/fpx(xi)
    Ea=np.abs(xi1-xi)
    print(i,xi1,Ea)
    if Ea<eps_max: break
    xi=xi1
  return

def NewtonRaphson_Mod(fx,fpx,fppx,xi,nit,eps_max):
  for i in range(nit):
    xi1=xi-(fx(xi)*fp(xi))/(fp(xi)**2-fx(xi)*fppx(xi))
    Ea=np.abs(xi1-xi)
    print(i,xi1,Ea)
    if Ea<eps_max: break
    xi=xi1

  return

def Secante(fx,xi,xim1,nit,eps_max):
  for i in range(nit):
    xi1=xi-fx(xi)*(xim1-xi)/(fx(xim1)-fx(xi))
    Ea=np.abs(xi1-xi)
    print(i,xi,Ea)
    if Ea<eps_max: break
    xi=xi1
  return
```

```python
f1=lambda x,y: x**2+x*y-10
f2=lambda x,y: y+3*x*y**2-57
df1dx= lambda x,y: 2*x+y
df1dy= lambda x,y : x
df2dx= lambda x,y: 3*y**2
df2dy= lambda x,y: 1+6*x*y
```

```python
def NewtonRaphson_sys(f1,f2,xi,yi):
  ui=f1(xi,yi)
  vi=f2(xi,yi)
  dvdyi=df2dy(xi,yi)
  dudyi=df1dy(xi,yi)
  dvdxi=df2dx(xi,yi)
  dudxi=df1dx(xi,yi)
  jac = dudxi*dvdyi-dudyi*dvdxi
  xit= xi - (ui*dvdyi-vi*dudyi)/jac
  yit= yi - (vi*dudxi-ui*dvdxi)/ jac
  return xit,yit
```

```python
xi,yi=1.5,3.5
for i in range(6):
  xi,yi=NewtonRaphson_sys(f1,f2,xi,yi)
  print(xi,yi)
```
